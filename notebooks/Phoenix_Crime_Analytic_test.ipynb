```python
#Cleaning Up the ZIP code data column to display integers, instead of floats
#First we need to remove all of the NaNs in the data set
```

```python
zipCodesNoNan = crimeData.dropna()
```

```python
zipCodesNoNan
```

```python
zipCodesNoNan['ZIP'].describe()
```

```python
crimeData = zipCodesNoNan
```

```python
crimeData
```

```python
zipCodeInt = crimeData['ZIP']
```

```python
zipNoFloat = [int(x) for x in zipCodeInt]
```

```python
crimeData['ZIP'] = zipNoFloat
```
